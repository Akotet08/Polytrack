```python
def get_video_list(directory, video_extension):
    video_list = []
    for video in os.listdir(directory):
        if video.endswith(video_extension):
            video_list.append((video.split('.')[0]))
    
    video_list.sort()

    return video_list

def get_image_name(csv_name, extension):
    file_name = csv_name[:-4]
    img_name = file_name+str('_img')+str(extension)

    return img_name

def generateTrackNumber(csv_name):
    species = csv_name.split('_')[0]
    track_code = getTrackCode(csv_name.split('_')[1])
    
    species_code = getSpeciesCode(species)
    
    track_number = str(track_code)+str(0)+str(species_code)
    
    return int(track_number)

def getTrackCode(_track):
    return _track[0:9]

def getSpeciesCode(_species):
    if _species == 'honeybee':
        _species_code = 0
    elif _species == 'hoverfly':
        _species_code = 1
    elif _species == 'moth':
        _species_code = 2
    elif _species == 'wasp':
        _species_code = 3
    else:
        _species_code = 4
        
    return _species_code

def getSpeciesName(_track):
    speciesnum = getSpeciesNumber(_track)
    
    if speciesnum == 0: _species = 'honeybee' #blue - honeybee
    elif speciesnum == 1: _species = 'hoverfly' #orage - hoverfly
    elif speciesnum == 2: _species = 'moth' #green - moth
    elif speciesnum == 3: _species = 'wasp' # red- wasp
    else: _species = 'other'
    
    return _species

def get_sec(time_str):
    """Get Seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

def gethour(t):   
    return int(t[3:5])
    

def checkUniqueNumber(_track_number, _track_list):
    while _track_number in _track_list:
        _track_number += 10
        
    return _track_number

def cal_dist(x,y,px,py):
    edx = x - px
    edy = y - py
    error = np.sqrt(edx**2+edy**2)
    return error


def getTime(t):
    return t[3:5] + ':' + t[5:7]+ ':' + t[7:9]

def getDataPoint(t):
    return int(t[0:1])

def getDataPointCode(t):
    return int(t[0:3])

def getRecordDate(t):
    return int(t[1:3])

def partition(pred, iterable):
    trues = []
    falses = []
    for item in iterable:
        if pred(item):
            trues.append(item)
        else:
            falses.append(item)
    return trues, falses

def getFlowerCode(dcp, flower):
    _dcp = str(dcp)
    _flower = str(flower)
    _flower_code = _dcp+_flower.zfill(2)
    
    return int(_flower_code) 

def getdcp_code(_date, _dcp):
    str_date = str(_date).zfill(2)
    str_dcp =str(_dcp)
    dcp_code = str_dcp+str_date
    
    return int(dcp_code)
```

```python
def generateFlowerPosition(csv_name):
    fp = csv_name.split('_')[1]
  
    return int(fp)

def get_position_matrix_to_plot(_position_matrix, _dates_to_plot):
    _position_matrix_plot = [[] for i in range(len(_position_matrix))]
    for p in range(len(_position_matrix)):
        for t in range(len(_position_matrix[p])): 
            if getRecordDate(str(_position_matrix[p][t])) in _dates_to_plot:
                _position_matrix_plot[p].append(_position_matrix[p][t])
                
    return _position_matrix_plot

def getSpeciesNumber(_track):
    speciesnum = str(int(_track))[-1]
    
    return int(speciesnum)

def getPlotColor(_track):
    _track = str(_track)
    _species = getSpeciesNumber(_track)
    
    if _species == 0: _color = '#1f77b4' #blue - honeybee
    elif _species == 1: _color = '#ff7f0e' #orage - hoverfly
    elif _species == 2: _color = '#2ca02c' #green - moth
    elif _species == 3: _color = '#d62728' # red- wasp
    else: _color = '#7f7f7f'
    
    return _color
```
